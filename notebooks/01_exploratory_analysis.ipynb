```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[
    [
        "fog_z",
        "sentence_z",
        "flesch_z"
    ]
] = scaler.fit_transform(
    df[
        [
            "gunning_fog",
            "avg_sentence_length",
            "flesch_reading_ease"
        ]
    ]
)
```

```python
df["complexity_score"] = (

      df["fog_z"]

    + df["sentence_z"]

    - df["flesch_z"]
)
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    df["complexity_score"],
    bins=30
)

plt.xlabel("Complexity Score")
plt.ylabel("Count")
plt.title("Distribution of Complexity Score")

plt.show()
```

```python
df["structure_score"] = (

    df["figure_count"]

    + df["section_count"]

    + df["reference_count"]
)
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    df["structure_score"],
    bins=30
)

plt.xlabel("Structure Score")
plt.ylabel("Count")
plt.title("Distribution of Structure Score")

plt.show()
```
